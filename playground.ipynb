```python
import pandas as pd
```

```python
# length is not enough for correlation
df = pd.DataFrame({'num_of_order': [124567],
                   'total_sales': [10747680],
                   'num_of_user': [79829]})
display(df.corr())
df.plot.scatter(x='num_of_order', y='total_sales')
```

```python
df = pd.DataFrame({'year': [2019, 2019, 2019],
                   'month': [1, 2, 3],
                   'num_of_order': [26, 95, 161],
                   'total_sales': [2115.970, 8132.340, 12748.23],
                   'num_of_user': [26, 95, 158]})
display(df.loc[:,['num_of_order', 'total_sales', 'num_of_user']].corr()['num_of_user'][:-1])
df.plot.scatter(x='num_of_order', y='total_sales')
```
